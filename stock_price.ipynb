```python
from scipy.stats import pearsonr
Selected_feature=['Open', 'High', 'Low','Volume', 'Daily_Return',
       'Price_Change', 'Price_Range', 'MA_7', 'MA_30', 'MA_90',
       'Volatility_30', 'Day_of_Week', 'Month', 'Year', 'Quarter']

correlation={
    feature:pearsonr(df_2[feature],df_2['Price'])[0]
    for feature in Selected_feature
}
correlation_df=pd.DataFrame(list(correlation.items()),columns=['feature','pearson_correlation'])
correlation_df.sort_values(by='pearson_correlation',ascending=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression


x=df_2[['Low','High','Open','MA_7','MA_30','MA_90','Year','Price_Range','Daily_Return','Price_Change','Day_of_Week',]]
y=df_2['Price']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train_scaled,y_train)
```

```python
y_pred=model.predict(X_test_scaled)
print(y_pred)
```

```python
df_2['Price']
```

```python
from sklearn.metrics import mean_squared_error,r2_score

mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print(f'mean squared error  {mse}')
print(f'r2 score  {r2}')
```

```python
train_score=model.score(X_train_scaled,y_train)
test_score=model.score(X_test_scaled,y_test)


print(f'mean squred error = {mse}')
print(f'r2 score = {r2}')
print(f'train score = {train_score}')
print(f'test score = {test_score}')
```

```python
from sklearn.model_selection import cross_val_score
cv_score = cross_val_score(model,X_train_scaled,y_train,cv=5)
print(f'all 5 score = {cv_score}')
print(f'average cv score = {cv_score.mean()}')
```
