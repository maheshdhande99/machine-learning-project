```python
df_2= pd.get_dummies(df_2,columns=['ocean_proximity'],drop_first=True).astype(int)
```

```python
df_2.head()
```

```python
from sklearn.preprocessing import StandardScaler
col=['longitude', 'latitude', 'housing_median_age', 'total_rooms','total_bedrooms', 'population', 'households', 'median_income']
scaler = StandardScaler()

df_2[col]=scaler.fit_transform(df_2[col])
```

```python
df_2.tail()
```

```python
from scipy.stats import pearsonr


selected_feature =['longitude', 'latitude', 'housing_median_age', 'total_rooms','total_bedrooms', 'population', 'households', 'median_income','ocean_proximity_INLAND',
       'ocean_proximity_ISLAND', 'ocean_proximity_NEAR BAY',
       'ocean_proximity_NEAR OCEAN']

correlation = {
    feature:pearsonr(df_2[feature],df_2['median_house_value'])[0]
    for feature in selected_feature
}
correlation_df = pd.DataFrame(list(correlation.items()),columns=['feature','pearson_correlation'])
correlation_df.sort_values(by='pearson_correlation',ascending=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_2.drop('median_house_value',axis=1)
y=df_2['median_house_value']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model =LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score

r2=r2_score(y_test,y_pred)
r2

n = X_test.shape[0]
p = X_test.shape[1]
ad_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))
ad_r2
```
